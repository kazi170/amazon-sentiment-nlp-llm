```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
sns.set(style="whitegrid")
```

```python
df=pd.read_csv("../data/raw/Amazon_review_dateset.csv",encoding='ISO-8859-1')
```

```python
#Checking what columns are available
df.columns
df.info()
df.head(5)
```

```python
#Lets only keep columns that we need and remove null values

df=df[['name', 'brand', 'reviews.date', 'reviews.rating', 'reviews.text']]
df=df.dropna(subset=['reviews.date', 'reviews.rating', 'reviews.text'])
```

```python
#Format Dates
df['reviews.date']=pd.to_datetime(df['reviews.date'],errors='coerce')
df = df.dropna(subset=['reviews.date'])

#plot Ratings to see how its distributed 

plt.figure(figsize=(8,5))
sns.countplot(x='reviews.rating',data=df,palette='viridis')
plt.title('Distribution of Review Rating')
plt.xlabel("Star Rating")
plt.ylabel("Count")
plt.show()


```

```python
#Loading pipelines 
from transformers import pipeline
# CardiffNLP RoBERTa: returns Positive / Neutral / Negative
model_roberta=pipeline("sentiment-analysis",model="cardiffnlp/twitter-roberta-base-sentiment")

#BMLPtown BERT: Returns star rating from 1 to 5
model_nlptown=pipeline("sentiment-analysis",model="nlptown/bert-base-multilingual-uncased-sentiment")

```

```python
#Taking a sample 
sample_reviews = df['reviews.text'].dropna().sample(5, random_state=42).tolist()

#comparing two models with sample reviews
for i, review in enumerate(sample_reviews):
    print(f"\n Review#{i+1}:{review[:300]}...")
    print("CardiffNLP RoBERTA Sentiment:", model_roberta(review))
    print("NLPtown Star Rating:", model_nlptown(review))
```

```python
#Using ROberta since its simpler to do sentiment analysis and see which products are getting negative reviews
label_map={'LABEL_0':'Negative','LABEL_1':'Neutral','LABEL_2': 'Positive'}

df_sample = df[['name','reviews.text', 'reviews.date', 'reviews.rating']].dropna().sample(10000, random_state=1).copy()


from tqdm import tqdm
tqdm.pandas()

def get_roberta_sentiment(review):
    try:
        result=model_roberta(review)[0]
        return label_map.get(result['label'],"Unknown")
    except:
        return "error"

df_sample['roberta_sentiment']=df_sample['reviews.text'].progress_apply(get_roberta_sentiment)

print(df_sample)
```

```python
#Check count of review sentiment
sentiment_count=df_sample['roberta_sentiment'].value_counts().reindex(['Positive','Neutral','Negative'])

sentiment_by_product=df_sample.groupby(['name','roberta_sentiment']).size().unstack(fill_value=0)

                                       
```

```python
#Sentiment distribution bar chart
plt.figure(figsize=(6,4))
sns.barplot(x=sentiment_count.index,y=sentiment_count.values)
plt.title('Distribution of Review Statement')
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
#creating a month column
df_sample['review_month']=df_sample['reviews.date'].dt.to_period('M')

#grouping them 
sentiment_trend=df_sample.groupby(['review_month','roberta_sentiment']).size().unstack(fill_value=0)
print(sentiment_trend)
#plot sentiment Trend over time 
plt.figure(figsize=(10,5))
sentiment_trend.plot(kind='line',marker='o')
plt.title('Sentiment over Time')
plt.ylabel('Number of Reviews')
plt.xlabel('Month')
plt.legend(title="Sentiment")

plt.tight_layout()



plt.show()

```

```python
#Highlighting top 10 negative sentiment products

#creating a shorter name 
df_sample['short name']=df_sample['name'].str.split().str[:4].str.join(" ")

sentiment_summary = df_sample.groupby('short name')['roberta_sentiment'].value_counts().unstack().fillna(0)

sentiment_summary['# Negative']=sentiment_summary['Negative'].astype(int)


#aggregate 
sentiment_cols = ['Negative', 'Neutral', 'Positive']

sentiment_summary['% Negative'] = (
    sentiment_summary['Negative'] / sentiment_summary[sentiment_cols].sum(axis=1) * 100
).round(1)

top_negative=sentiment_summary.sort_values('% Negative', ascending=False).head(10)
print(top_negative)

#plot the products with the most negative sentiment 
plt.figure(figsize=(6,5))
top_negative['% Negative'].plot(kind='barh',color='crimson')
plt.title('Top 10 Negative Sentiment Products')
plt.xlabel('% Negative')
plt.ylabel('Product Name')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

```

```python
#summarize negative reviews 
#High %Negative product reviews
#Identify products where dissatisfaction is unusually high relative to their review base.

high_percent_negative = sentiment_summary[
    (sentiment_summary['% Negative'] >= 10)   # ensure some review volume
].sort_values('% Negative', ascending=False)

#Display some of the negative reviews for the high %negative reviews low volume
for prod in high_percent_negative.index:
    sample_reviews=df_sample[(df_sample['short name']==prod) & (df_sample['roberta_sentiment']=='Negative')]['reviews.text'].tolist()
    print(f"\n {prod}-{len(sample_reviews)} negative reviews")

    for r in sample_reviews[:4]:
        print(".",r[:400], "...\n")

#Understand themes in frequent customer complaints for high-impact products
#High Negative Count (Volume for summarization)

# Filter products with at least 25 negative reviews and at least 5% negative rate
top_candidates=sentiment_summary[(sentiment_summary['Negative']>=25) & (sentiment_summary['% Negative']>=5)].sort_values('% Negative',ascending= False).head(2)

target_product1=top_candidates.index[0]
target_product2=top_candidates.index[1]

#Extract negative reviews for that product
neg_reviews_product1=df_sample[(df_sample['short name']==target_product1) & (df_sample['roberta_sentiment']=='Negative')]['reviews.text'].tolist()
neg_reviews_product2=df_sample[(df_sample['short name']==target_product2) & (df_sample['roberta_sentiment']=='Negative')]['reviews.text'].tolist()


#need to break down the reviews in chunk sizes for the LLM 
def chunk_reviews(reviews,chunk_size=5):
    chunks=[]
    for i in range(0,len(reviews),chunk_size):
        chunk=reviews[i:i+chunk_size]
        combined=' '.join(chunk)
        chunks.append(combined)
    return chunks


chunks_prod1=chunk_reviews(neg_reviews_product1)
chunks_prod2=chunk_reviews(neg_reviews_product2)



```
